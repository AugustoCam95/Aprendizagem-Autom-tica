```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
data = np.loadtxt("ex1data1.txt", dtype = float, delimiter = ',')
```

```python
X = data[:,0]
Y = data[:,1]
```

```python
x = np.ones((data.shape[0],2))
x[:,1] = X
p = x[2,:]
print(p[1])
```

```python
w = np.zeros((2,1))
print(w[0])
print(len(w))
```

```python
#Setar o alpha
alpha = 0.001
n_epochs = 1000
n = data.shape[0]
eqm = []
```

```python
for i in range(n_epochs):
    ind = np.random.permutation(n)
    x = x[ind,:]
    y = Y[ind]
    erro_total = 0
    for j in range(n):
        y_barra = np.matmul(x[j,:],w)
        erro = y[j] - y_barra[0]
        erro_total = erro_total + erro*erro
        w[0] = w[0] + (alpha*erro)
        w[1] = w[1] + (alpha*erro*x[j,1])
    eqm.append((erro_total/n))

eixo_x = range(n_epochs,0,-1)


plt.plot(eixo_x,eqm)
plt.show()
```
